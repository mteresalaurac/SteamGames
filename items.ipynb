```python
import json
import ast
import gzip
import pandas as pd

#Abro el archivo de User ITEMS
#Abro el archivo de User ITEMS
data = []
with gzip.open('Archivos_Originales/users_items.json.gz','rb') as f:
    for line in f:
        data.append(ast.literal_eval(line.decode("latin-1")))
items=pd.DataFrame(data)
f.close()

items
```

```python
#Me fijo que contiene el dataframe adentro
items.head(5)
```

```python
#Items > Voy a ver que tengo en la columna de items. Tengo una lista, y esa lista tiene varios diccionarios: varios items:
items['items'].iloc[5]
```

```python
#Desanido el diccionario que está en items: en la tercer columnas que tambien se llama items.. 
#Voy a generar un item por cada user ID usuario (antes tenia el user id, y asociados todos los items a la vez)

items_final=items.explode('items')
items_final.head(5)
```

```python
#Evaluo que tengo en la columna de items. Dentro del diccionario aparecen 4 conceptos: item_id; item_name; playtime_forever y playtime 2 weeks.

items_final["items"].iloc[0]
```

```python
#Voy a resetear el indice de items_final 
#Cuando pude abrir el primer diccionario que tiene el file de items, logre que se separen los distintos items que pertenecen a un mismo user id.
#Voy a asignarle un indice a cada fila

items_final.index = range(len(items_final))

items_final.head(5)
```

```python
#Voy a desanidar cada item
#Creo un df a partir de la lista de elementos anidados en la review.

items_desanidados = pd.json_normalize(items_final["items"]) # Le paso la columna de reviews. Genera una lista d elementos anidados. Armo un DF, con cada clave del diccionario dentro de items como una columna
items_desanidados
```

```python
import pandas as pd
#Asocio cada review con el user id correspondiente. 
items_clean = pd.concat([items_final.drop(columns= ['items']), items_desanidados], axis=1) #concateno el df extendido (con una fila por review de usuario) con el df con una columna por key de review

items_clean.head(5)
```

```python
#Empiezo a limpiar el archivo
#Voy a eliminar las filas con nulos.
#Con la especificación how="all" voy a indicarle que elimine las filas donde TODOS los elementos son nulos.
items_clean=items_clean.dropna(how="all")

print(items_clean.shape) # Mismo tamaño pq no hay nulos totales >  no hace falta
```

```python
#Voy a eliminar las reviews que estan duplicadas
items= items_clean.duplicated()

#print(sum(duplicados)) #Devuelve 874, asique voy a eliminar esos duplicados.

items_sinduplicados=items_clean.drop_duplicates()

print(items_sinduplicados.shape) #Elimino algunas filas
```

```python
#Elimino columnas que no voy a necesitar y me quedo un df mas corto.
#user_id y steam_id son lo mismo por lo tanto elimino la columna

items_final= items_sinduplicados.drop(columns=["items_count","steam_id","user_url","playtime_2weeks"]) #SteamId es lo mismo que user_id, por eso lo elimino

print(items_final.shape) #Veo que achique las dimensiones en este caso.
```

```python
#Reviso como quedó el df luego de la limpieza. 

items_final.head(3)
```

```python
#Modifico el nombre de la columna con las horas jugadas para trabajar mas facil con la nomenclatura
items_final.rename(columns={'playtime_forever': 'tiempo_min'}, inplace=True)

#Como el tiempo esta en minutos, lo paso a horas y luego convierto nuevamente el nombre de la columna

items_final['tiempo_min'] = items_final['tiempo_min'].div(60)

items_final.rename(columns={'tiempo_min': 'horas'}, inplace=True)

```

```python
#Guardo items_final como parquet para utilizarlo en las funciones si necesito mas espacio
items_final.to_parquet("Archivos_Comprimidos/items.parquet",index=False)
```
